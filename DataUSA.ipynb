# 3. Write python code that will give us the response time for each mail piece that has been responded to.

response time = response date - mail date

**Note**: Response time is in the form of #_of_months/#_of_days/#_of_years that have past

```python
df3 = pd.merge(df_cont, df_mail, how='inner')
df3 = pd.merge(df2, df_re, how='inner')
```

```python
def get_response_time(times):
  # get mail times
  time_m = times[0].split("/")
  day_m = int(time_m[0])
  month_m = int(time_m[1])
  year_m = int(time_m[2])

  # get resp times
  time_r = times[1].split("/")
  day_r = int(time_r[0])
  month_r = int(time_r[1])
  year_r = int(time_r[2])

  return str(abs(day_r-day_m)) +"/"+ str(abs(month_r-month_m)) +"/"+ str(abs(year_r-year_m))
```

```python
df3['RESPONSE_TIME'] = df3[['MAIL_DATE','RESP_DATE']].apply(get_response_time, axis=1) #passes a Series object, row-wise
```

# 8. Write python code that will create the FIRST_NAME and LAST_NAME column from the value found in the NAME column.
* Assume that the NAME column is [FIRST_NAME] [LAST_NAME] separated with a space
* Assume every NAME contains first and last name. Invalid names have already been
filtered out

```python
def get_fname(name):
  names = name.split(" ")
  return names[0]

def get_lname(name):
  names = name.split(" ")
  return names[1]
```

```python
df['FIRST_NAME'] = df['NAME'].apply(get_fname)
df['LAST_NAME'] = df['NAME'].apply(get_lname)
```
